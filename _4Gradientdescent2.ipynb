Find the correlation between the scores of two subjects using gradient descent algorithm, compare the previous cost and current cost between each iteration when it is within certain threshold( stop when costs are similar-> using math.isclose function with 1e-20 threshold). Using sklearn.linear_model to find coefficient and intercept, and compare them with those generated by gradient descent algorithm.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import math
```

```
def predict_using_sklearn():
  data = { 'name' : ['david','laura','sanjay','wei','jeff','aamir','venkat','virat','arthur','paul'] ,
          'math' : [92,56,88,70,80,49,65,35,66,67] ,
          'cs' : [98,68,81,80,83,52,66,30,68,73] 
          }
  df = pd.DataFrame(data, columns= ['name','math','cs'])
  reg=LinearRegression()
  reg.fit(df[['math']],df.cs)
  return reg.coef_, reg.intercept_
```

```
def gradient_descent(x,y):
  m_curr=0
  b_curr=0
  iterations=1000000
  n=len(x)
  learning_rate=0.0002
  prev_cost=0
  for i in range(iterations):
    y_pred = m_curr*x + b_curr
    cost= 1/n * sum([val**2 for val in (y - y_pred)])
    md = -(2/n)*sum(x*(y-y_pred))
    bd = -(2/n)*sum(y-y_pred)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd 
    if math.isclose(cost, prev_cost, rel_tol=1e-20):
            break
    prev_cost = cost
    print ("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost, i))

  return m_curr, b_curr

```

```
data = { 'name' : ['david','laura','sanjay','wei','jeff','aamir','venkat','virat','arthur','paul'] ,
          'math' : [92,56,88,70,80,49,65,35,66,67] ,
          'cs' : [98,68,81,80,83,52,66,30,68,73] 
          }
df = pd.DataFrame(data, columns= ['name','math','cs'])
x = np.array(df.math)
y = np.array(df.cs)
m, b = gradient_descent(x,y)
print("Using gradient descent function: Coef {} Intercept {}".format(m, b))

m_sklearn, b_sklearn = predict_using_sklean()
print("Using sklearn: Coef {} Intercept {}".format(m_sklearn,b_sklearn))
```
